```
import pandas as pd
import numpy as np

# Parameters for wave generation
n_samples_normal = 1000  # Number of normal signals
signal_length = 100  # Length of each signal

# Function to generate normal wave data (sinusoidal)
def generate_normal_wave(n_samples, signal_length):
    time = np.linspace(0, 10, signal_length)
    signals = []
    for _ in range(n_samples):
        freq = np.random.uniform(1, 5)  # Random frequency for each signal
        amplitude = np.random.uniform(0.5, 1.5)  # Random amplitude
        signal = amplitude * np.sin(2 * np.pi * freq * time)
        signals.append(signal)
    return np.array(signals)

# Generate normal rail data
normal_data = generate_normal_wave(n_samples_normal, signal_length)

# Create clear column names for the signals
column_names = [f'signal_{i}' for i in range(signal_length)]

# Create a DataFrame with the generated normal data
df_normal = pd.DataFrame(normal_data, columns=column_names)
df_normal['label'] = 0  # Label 0 for normal rail data

# Save the normal dataset to CSV
df_normal.to_csv('acoustic_wave_normal.csv', index=False)

print("Normal dataset saved as 'acoustic_wave_normal.csv'")
```

```
# Parameters for wave generation
n_samples_cracked = 1000  # Number of cracked signals

# Function to generate cracked wave data (sinusoidal with noise)
def generate_cracked_wave(n_samples, signal_length):
    time = np.linspace(0, 10, signal_length)
    signals = []
    for _ in range(n_samples):
        freq = np.random.uniform(1, 5)  # Random frequency
        amplitude = np.random.uniform(0.5, 1.5)  # Random amplitude
        noise = np.random.normal(0, 0.2, signal_length)  # Add noise to simulate cracks
        signal = amplitude * np.sin(2 * np.pi * freq * time) + noise
        signals.append(signal)
    return np.array(signals)

# Generate cracked rail data
cracked_data = generate_cracked_wave(n_samples_cracked, signal_length)

# Create a DataFrame with the generated cracked data
df_cracked = pd.DataFrame(cracked_data, columns=column_names)  # Use the same column names
df_cracked['label'] = 1  # Label 1 for cracked rail data

# Save the cracked dataset to CSV
df_cracked.to_csv('acoustic_wave_cracked.csv', index=False)

print("Cracked dataset saved as 'acoustic_wave_cracked.csv'")
```

```
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

# Load the CSV data (assuming it's acoustic wave data for normal rail tracks)
data = pd.read_csv('/content/acoustic_wave_normal.csv')  # Replace with your file name

# Feature scaling (important for Isolation Forest)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)  # Scale the data

# Train the Isolation Forest model using normal data (normal rail track data)
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # Adjust contamination for anomaly ratio
iso_forest.fit(X_scaled)

# Predict: -1 indicates an anomaly, 1 indicates normal
predictions = iso_forest.predict(X_scaled)

# Convert predictions: -1 (anomaly/crack), 1 (normal)
data['prediction'] = predictions

# Count the predictions
anomalies = np.sum(data['prediction'] == -1)
normals = np.sum(data['prediction'] == 1)

print(f"Detected {anomalies} anomalies (potential rail cracks) and {normals} normal signals.")

# Optional: If you have cracked rail data for testing, load and evaluate
cracked_data = pd.read_csv('/content/acoustic_wave_cracked.csv')  # Test data for cracked rails
cracked_scaled = scaler.transform(cracked_data)  # Scale the data
cracked_predictions = iso_forest.predict(cracked_scaled)

# If you want to evaluate:
y_true = [1] * len(cracked_data)  # Assuming all data is cracked
y_pred = [-1 if p == -1 else 1 for p in cracked_predictions]  # Convert to 1 for cracked, -1 for normal

#Print a report of the performance
print("Classification Report:\n", classification_report(y_true, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
```

```
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE

# Load dataset from a CSV file (replace 'your_dataset.csv' with your actual dataset file)
data = pd.read_csv('your_dataset.csv')

# Assume 'Label' is the target column (0: No Crack, 1: Crack Detected) and the rest are features
X = data.drop('Label', axis=1)
y = data['Label']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the data (scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Handle imbalanced data using SMOTE (optional, if dataset is imbalanced)
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Initialize the model
model = LogisticRegression()

# Hyperparameter tuning using GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100], 'solver': ['lbfgs', 'liblinear']}
grid = GridSearchCV(model, param_grid, refit=True, verbose=2)
grid.fit(X_train_resampled, y_train_resampled)

# Get the best hyperparameters
best_model = grid.best_estimator_

# Evaluate using Cross-Validation (for robust evaluation)
cv_scores = cross_val_score(best_model, X_train_resampled, y_train_resampled, cv=5)
print(f"Cross-validation Accuracy: {cv_scores.mean() * 100:.2f}%")

# Train the model with best hyperparameters
best_model.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred = best_model.predict(X_test_scaled)

# Calculate and display model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Evaluate using precision, recall, and F1-score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

# Classification report (more detailed metrics)
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix to visualize model performance
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# ROC Curve and AUC score to measure the tradeoff between true positive and false positive rates
y_proba = best_model.predict_proba(X_test_scaled)[:, 1]  # Get probability estimates for the positive class
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc = roc_auc_score(y_test, y_proba)

plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="best")
plt.show()

# Step-by-step predictions (output for each test sample)
print("\nStep-by-step detection of rail cracks:")
for i, pred in enumerate(y_pred):
    if pred == 1:
        print(f"Test Sample {i+1}: Crack Detected")
    else:
        print(f"Test Sample {i+1}: No Crack Detected")
```
